```python
import pandas as pd

data = pd.read_excel('./raw_data.xlsx')
pd.set_option('display.max_columns', 50)

data.head()
```

```python
def deduplicate_name(row):
    if len(data[data["nom"] == row["nom"]]) > 1:
        row["nom"] = row["nom"] + " " + row["commune"]
    return row

clean = data[["nom", "commune"]].apply(deduplicate_name, axis=1)
data["nom"] = clean["nom"]

data.to_excel('./clean_data.xlsx', index=False)
```
