```python
import pandas as pd

data = pd.read_excel('./raw_data.xlsx')
pd.set_option('display.max_columns', 50)

data.head()
```

```python
def deduplicate_name(name):
    if len(data[data["nom"] == name]) > 1:
        return name + " " + data[data["nom"] == name]["commune"].iloc[0]
    else:
        return name

data["nom"] = data["nom"].agg(deduplicate_name)

data.to_excel('./clean_data.xlsx', index=False)
```
